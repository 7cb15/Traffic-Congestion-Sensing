```python
# ============================================================================
IMAGE_DIR = "../out_test"
VIDEO_SOURCE = "../videos/sample.mp4"
IMAGE_FOLDER = "../gopro_cropped"
SHAPE = (1500, 2000)  # HxW
EXIT_PTS = np.array([
    [[785,800], [755,650], [880,625], [950,760]],
    [[960,1410], [900,1190], [1050,1190], [1145,1390]]
])
FILTER_MASK = np.array([
    [[940,1460],[620,400],[725,390],[1350,1420]]
])
# ============================================================================
```

### Filtering/Foreground 

```python
plt.figure(figsize=(12,5))

# Create base to fill
base = np.zeros(SHAPE + (3,), dtype='uint8')

# Create exit zones
exit_mask = cv2.fillPoly(base, EXIT_PTS, (255, 255, 255))[:, :, 0]
plt.subplot(1,2,1)
plt.title('exit zones')
plt.imshow(exit_mask)

# Create filter mask
filter_mask = cv2.fillPoly(base, FILTER_MASK, (255, 255, 255))[:, :, 0]
plt.subplot(1,2,2)
plt.title('mask for filter')
plt.imshow(filter_mask)
```

### Foreground Detection

```python
def train_bg_subtractor(inst, cap, num=100):
    '''
        BG substractor need process some amount of frames to start giving result
    '''
    print ('Training BG Subtractor...')
    i = 0
    while True:
        ret, frame = cap.read() 
        if frame is None:
            break
#     for frame in cap:
        inst.apply(frame, None, 0.001)
        i += 1
        if i >= num:
            return cap
```

```python
# background subtractor
# will increase 'history/num' when running on full video file
bg_subtractor = cv2.createBackgroundSubtractorMOG2(
        history=100, detectShadows=True)

cap = cv2.VideoCapture('../videos/sample.mp4')

#this eliminates first n records
train_bg_subtractor(bg_subtractor, cap, num=100)

frame_number = -1
while True:
    ret, frame = cap.read()
    if frame is None:
        log.error("Frame capture failed, stopping...")
        break
        
    frame_number += 1
    utils.save_frame(frame, "../out_test/frame_%04d.png" % frame_number)

    fg_mask = bg_subtractor.apply(frame, None, 0.001)

    utils.save_frame(fg_mask, "../out_test/fg_mask_%04d.png" % frame_number, flip=False)    
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title('Original')
img = cv2.imread('../out_test/frame_0200.png')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Eliminated Background')
img = cv2.imread('../out_test/fg_mask_0200.png')
plt.imshow(img)
```

### Filters after background subtraction
after these are curated we will add changes to filter script for better detection

```python
img = cv2.imread('../out_test/fg_mask_0200.png')
```

```python
plt.figure(figsize=(12,15))
plt.suptitle('Filtering Process', fontsize=18)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (1,1))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (8,8))
kernel3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))

# Original image
plt.subplot(3,2,1)
plt.title('Eliminated Background')
plt.imshow(img)

# Filter out road
isolated = cv2.bitwise_or(img, img, mask=filter_mask)
plt.subplot(3,2,2)
plt.title('Isolate Road')
plt.imshow(isolated)

# Fill any small holes
closing = cv2.morphologyEx(isolated, cv2.MORPH_CLOSE, kernel)
plt.subplot(3,2,3)
plt.title('Closing technique')
plt.imshow(closing)

opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel2)
plt.subplot(3,2,4)
plt.title('Opening technique')
plt.imshow(opening)

# Dilate to merge adjacent blobs
dilation = cv2.dilate(opening, kernel3, iterations=2)
plt.subplot(3,2,5)
plt.title('Dilation technique')
plt.imshow(dilation)

# threshold
dilation[dilation < 240] = 0
plt.subplot(3,2,6)
plt.title('Threshold technique')
plt.imshow(dilation)

# plt.tight_layout()
plt.show()
```

### After adjustments made to filter_mask function (in pipeline.py), run the following

```python
# processing pipline for programming conviniance
pipeline = PipelineRunner(pipeline=[
    ContourDetection(bg_subtractor=bg_subtractor,
                     save_image=True, image_dir=IMAGE_DIR),
    # we use y_weight == 2.0 because traffic are moving vertically on video
    # use x_weight == 2.0 for horizontal.
    VehicleCounter(exit_masks=[exit_mask], y_weight=2.0),
    Visualizer(image_dir=IMAGE_DIR),
    CsvWriter(path='./', name='report.csv')
], log_level=logging.DEBUG)

bg_subtractor = cv2.createBackgroundSubtractorMOG2(
        history=100, detectShadows=True)

cap = cv2.VideoCapture('videos/sample.mp4')

train_bg_subtractor(bg_subtractor, cap, num=100)

_frame_number = -1
frame_number = -1

while True:
    ret, frame = cap.read()
    if frame is None:
        log.error("Frame capture failed, stopping...")
        break
        
    # real frame number
    _frame_number += 1

    # skip every 2nd frame to speed up processing
#     if _frame_number % 2 != 0:
#         continue

    # frame number that will be passed to pipline
    # this needed to make video from cutted frames
    frame_number += 1

    pipeline.set_context({
        'frame': frame,
        'frame_number': frame_number,
    })
    pipeline.run()

print(_frame_number, frame_number)
```
