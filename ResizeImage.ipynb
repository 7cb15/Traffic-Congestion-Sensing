```python
from PIL import Image
import os
from matplotlib.pyplot import imshow
import numpy as np

%matplotlib inline
```

```python
# test
image = Image.open('sample_images/G0137246.JPG')
imshow(np.asarray(image))
```

```python
cropped = image.crop((1500, 1500, 3500, 3000))
imshow(np.asarray(cropped))
```

```python
def cropImage(imagepath, outputpath):
    img = Image.open(imagepath)
    crop_img = img.crop((1500, 1500, 3500, 3000))
    crop_img.save(outputpath)
```

```python
path = 'sample_images'
output = 'sample_images_output'
directory = os.fsencode(path)
outdir = os.fsencode(output)

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".JPG"):
        imageSource = os.path.join(os.fsdecode(directory), filename)
        imageOut = os.path.join(os.fsdecode(outdir), filename)
        cropImage(imageSource, imageOut)
```
